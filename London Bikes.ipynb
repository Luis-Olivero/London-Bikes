```python
# targeting specific columns for use
new_cols_dict = {
    'timestamp':'time',
    'cnt':'count', 
    't1':'temp_real_C',
    't2':'temp_feels_like_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

# Changing target columns names
bikes.rename(new_cols_dict, axis=1, inplace=True)
```

```python
# Converting humidity values to percentage
bikes.humidity_percent = bikes.humidity_percent / 100
```

```python
# Converting season column header from integers to written values creating a season dictionary
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}

# Same process for weather dictionary
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}

# Converting season column data type to string
bikes.season = bikes.season.astype('str')
# mapping the values 0-3 to the actual written seasons
bikes.season = bikes.season.map(season_dict)

# Convert weather column data type to string
bikes.weather = bikes.weather.astype('str')
# mapping the values 0-3 to the actual written weathers
bikes.weather = bikes.weather.map(weather_dict)
```

```python
# Visual check our dataframe to see if the mappings are functioning properly
bikes.head()
```

```python
# Specify the file path where you want to save the Excel file
excel_file_path = r"C:\Users\Luisa\OneDrive\Documents\Data Practice\London Bike Sharing"

# Define the file name and sheet name
excel_file_name = "london_bikes_final.xlsx"
sheet_name = "Data"

# Construct the full file path including the file name
full_file_path = excel_file_path + "\\" + excel_file_name

# Export the DataFrame to an Excel file
bikes.to_excel(full_file_path, sheet_name=sheet_name)

# Print a message to confirm the export
print(f"DataFrame successfully exported to '{full_file_path}'")
```
